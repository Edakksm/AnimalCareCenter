## Task 2️⃣: Choose between conflicting sessions

One of the most powerful features of a LLM using AI is that it can evaluate multiple things with the same instructions, rather than you having to copy and paste the same instructions over and over again. Let's try it out with conference sessions.

At AMIA Amplify, some of the hardest choices happen when several good sessions are scheduled at the same time. Let's have AI evaluate a few different sessions to see which one we should attend.

First, we'll create a small table (also called a data frame) of AMIA sessions.
- Each row will be one session
- The `session_name` column has the name of the session

```python
# Below is a small list of AMIA sessions.
# Each { } holds information about one session.

# STEP 1:
# Change *** to a session you might be excited to attend at AMIA Amplify to add it to the table.
# Keep the quotation marks.

sessions = [
    {"session_name": "CI24 Inboxology: The New Paradigm of Message-based Patient Care"},
    {"session_name": "CI26 Generative AI Enters the Hospital"},
    {"session_name": "TRI17 From Clusters to Clouds: Designing Secure, HIPAA-Ready Infrastructure"},
    {"session_name": "TRI24 Seeing Is Believing: Imaging & Vision AI"},
    {"session_name": "***"}
]

# We are naming the table sessions_table.
sessions_table = pd.DataFrame(sessions)

# Print the table.
sessions_table
```

Nice! Now we have a table of sessions.

Notice the numbers on the far left: `0`, `1`, `2`, `3`, `4`.

These are the **row numbers** (of note- it starts at row 0!).

In the next step, change the chosen row to each row number to get AI's recommendation about the session.


```python
# STEP 1:
# Choose which row you want to look at.
# Try changing this number and running the cell again.
chosen_row = ***

# Pull out the session from that row.
each_session_row = sessions_table.iloc[chosen_row]["session_name"]

#All this is the same as before now, except we're changing the prompt to add the session name and number.
system_prompt = "You are a helpful AMIA Amplify conference assistant."

user_prompt = f"""
Here is my profile:
{my_profile}

Here is one session I am considering:
Session number: {each_session_row}

Should I attend this session?

Respond exactly in this format:
Recommendation: attend / maybe / skip
Reason: <one sentence>
"""

results = client.chat.completions.create(
    model="gemini-2.5-flash",
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt}
    ],
)

print(results.choices[0].message.content)
```

You reviewed all the sessions, but imagine if you wanted to go through *all* the sessions at AMIA- you'd be here forever changing that number!

Let's have AI through the table one row at a time automatically. That is called a **loop**.

## Task 3️⃣: Review all the sessions

A loop tells the computer:

➡️ “Repeat the same steps again and again.”

In this task, the loop will go through the session table **one row at a time**. For each row, it will:

1. pull out the session name  
2. ask the AI whether that session seems like a good fit  
3. print the answer


Here is the line of code that accoplishes the loop in this task:

`for i, row in sessions_table.iterrows():`


- `for` starts the loop
- `row` means “the current row in the table”
- `iterrows()` tells Python to go through the table one row at a time


```python
# System prompt
system_prompt = "You are a helpful AMIA Amplify conference concierge."

# Start the loop.
# This tells Python to go through the table one row at a time.
for i, row in sessions_table.iterrows():

    # Pull out the session name from the current row.
    each_session_name = row["session_name"]

    # User prompt
    user_prompt = f"""
Here is my profile:
{my_profile}

Here is one session I am considering: {each_session_name}

Should I attend this session?

Respond exactly in this format:
Recommendation: attend / maybe / skip
Reason: <one sentence>
"""

    # Send the prompts to the model
    results = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
    )

    # Print the session we are reviewing + the model's answer
    print(each_session_name)
    print(results.choices[0].message.content)
    print()
```

Nice work! You just used a loop.


## Task 4️⃣: Save your shortlist as a CSV file

So far, we’ve been printing one recommendation at a time.

That is helpful for checking whether the prompt works, but if we want to **keep** the results, we need to store them somewhere.

A **CSV file** (comma-separated values) is a simple file that stores data in a table format, like a spreadsheet. You can open it in Excel, Google Sheets, or a text editor.


```python
# Let's create an empty container to store outputs.
# As the loop runs, we will add one new item to each list.
session_list = []
ai_output = []

# This is all the same from the last step.
system_prompt = "You are a helpful AMIA Amplify conference concierge."

for i, row in sessions_table.iterrows():
    each_session_row = row["session_name"]

    user_prompt = f"""
Here is my profile:
{my_profile}

Here is one session I am considering: {each_session_row}

Should I attend this session?

Respond exactly in this format:
Session:
Recommendation: attend / maybe / skip
Reason: <one sentence>
"""

    results = client.chat.completions.create(
        model="gemini-2.5-flash",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
    )

    # Select the model's output.
    output_text = results.choices[0].message.content

    # Add each new item to the containers.
    ai_output.append(output_text)
    session_list.append(each_session_row)

# Each item becomes one column in the table.
results_table = pd.DataFrame({
    "Session": session_list,
    "Model Output": ai_output
})

# STEP 1: Change *** to a name that you want to save your file under it
# Save the table as a CSV file.
output_filename = "***.csv"
results_table.to_csv(output_filename, index=False)
```

# **🚨 Common Pitfalls in LLM API Implementation**
Building with LLM APIs is exciting... until your workflow suddenly stops working at 2 AM!
<by>Here are some of the most common pitfalls when you work with LLM APIs and how to avoid them.

##**⚙️ Technical Constraints**
LLM APIs limit how quickly (requests per minute, RPM) and how much text (tokens per minute, TPM) you can process
<br>Too much data too fast? Your requests may fail, slow down, or return incomplete results
###What can you do to address this?
* Process data in smaller batches rather than all at once
* Choose models that can handle the amount of text you are working with

##**🎯 Reliability and reproducibility**
Same prompt → variable results sometimes
<br>Minimum prompt changes → Surprisingly different outputs
<br>No validation → unnoticed prompt/model changes and model drift
###What can you do to address this?
* Use structured prompts and consistent output formats
* Review outputs regularly and refine prompts as needed
* Just like your favorite Word document, keep version control
* Benchmark against gold standards whenever possible

##**🛠️ Operational and workflow integration**
One broken step can stop an entire workflow during long-running jobs.
<br>Token-based pricing → large datasets can become expensive very quickly
<br> Confident-sounding AI outputs are not always correct
###What can you do to address this?
* Add checkpoints to help your workflow recover after errors
* Monitor real-time cost and token utilization so that you are not surprised when you get the bill!
* Keep the tasks focused and well defined

##**🩺 PHI considerations**
Not all APIs are safe for PHI! Identifiable data is protected by HIPAA. Using an API that is not safe for PHI will send that data to the company that owns the LLM and potentially lead to that information being used for training new models.
<br>**❗In this workshop, your Gemini API key and Google Co-lab are NOT HIPAA-safe. Do NOT enter PHI or identifiable patient data!**

###What can you do to address this?
 * If your institution offers an API key that is safe for clinical data, you can use this.
 * If your institution does not offer an institutional API key, you can run a local model, which is a large language model that runs on your computer. Your computer does need to have certain specifications to run a model well.
